# Text → Top-3 Similar Reference Views

Standalone lookup tool — no dependency on `CameraReasoningSession`, the camera, or any iteration loop from `manual_chatgpt_loop.ipynb`. Type any text describing a view and see which reference views (`reference_views/view_descriptions.json`) it's closest to, via CLIP text embeddings.

```python
import sys, os
# Add project root so `camera_reasoning` is importable when running from notebooks/
sys.path.insert(0, os.path.abspath('..'))

from camera_reasoning.text_view_matcher import TextViewMatcher
```

## Build the index (once)

Loads from the cache if present (`reference_text_embeddings.pkl`), otherwise encodes `view_descriptions.json` from scratch and writes the cache.

```python
text_matcher = TextViewMatcher(
    descriptions_path="../reference_views/view_descriptions.json",
    cache_path="../reference_text_embeddings.pkl",
)
text_matcher.build_index()
```

## Look up any text

Edit `query_text` below and re-run.

```python
def top3_for_text(query_text: str):
    result = text_matcher.match(query_text, top_k=3)
    for m in result["top_matches"]:
        print(f"{m['rank']}. {m['view_label']:<20s} similarity={m['similarity']:.4f}")
    return result

query_text = "The current screenshot shows a foot bone structure viewed from a strong oblique angle with significant depth/perspective. The foot appears to be tilted, showing a side profile view. I can see the toes (phalanges) at the upper left, metatarsals in the middle, and ankle/tarsal region at the lower right. The view emphasizes the side/lateral profile of the foot with significant three-dimensional depth, showing the contours and relief of the bone structure rather than a flat top-down perspective."
top3_for_text(query_text)
```
